```python
%pylab inline 
%load_ext autoreload
%autoreload 2
from lxml import etree as Etree
import matplotlib
import matplotlib.pyplot as plt
import sys
import numpy as np
import subprocess
```

```python
# visulize the performance of the training algorithm.
def parseBlock(block):
    block = block.split('\t')
    img = [int(b.split('/')[0]) for b in block if b != ""]
    tag = [int(b.split('/')[1]) for b in block if b != ""]
    return img, tag

def visualizeExample(root, path):
    truth = root.findall(path+'/truth')[0].text
    tag = root.findall(path+'/tag')[0].text
    tag = tag.split('\n')[1]
    headline = truth.split('\n')[1]
    headline = headline.split(' ')
    H = int(headline[0])
    W = int(headline[1])
    (img, tag) = parseBlock(tag)
    (img, gt) = parseBlock(truth.split('\n')[2])
    gt = np.reshape(gt, (H, W), order='F')
    img = np.reshape(img, (H, W), order='F')
    tag = np.reshape(tag, (H, W), order='F')
    print 'image = '
    imshow(img)
    show()
    print 'tag = '
    imshow(tag)
    show()
    print 'truth = '
    imshow(gt)
    show()
name = 'result/ising.xml'
tree = Etree.parse(name)
root = tree.getroot()
pass_s = 'pass_2'
visualizeExample(root, pass_s + '/test/example_0')
score = root.findall(pass_s + '/score')
print score[-1].text
```

```python
# visualize mask of adaptive inference algorithm.
def visualizeMask(mask):
    mask = mask.split('\n')[1]
    mask = mask.split('\t')
    mask = [float(mk) for mk in mask if mk != '']    
    mask = np.reshape(mask, (128, 128), order='F')
    imshow(mask)
    show()

name = 'test/ising/lockdown/initIID/0/tc99999_c-2.9254443717194922/policy.xml'
tree = Etree.parse(name)
root = tree.getroot()
mask = root.findall('test/example/example_1/mask')[0].text
visualizeMask(mask)
```

```python
# accuracy vs. time.
def execute(cmd):
    ps = subprocess.Popen(cmd, shell=True,stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    output = ps.communicate()[0]
    return output

def linesToArray(lines):
    lines = lines.split('\n')
    return [float(line) for line in lines if line != '']
    
def loadAcc(path):
    acc = linesToArray(execute("for file in `ls -t %s`; do cat $file | grep '<accuracy>' -A 1 | tail -n 1; done" % path))
    time = linesToArray(execute("for file in `ls -t %s`; do cat $file | grep '<time>' -A 1 | tail -n 1; done" % path))
    (time, acc) = zip(*sorted(zip(time, acc), key=lambda x : x[0]))
    return (list(time), list(acc))

def loadAccWallclock(path):
    acc = linesToArray(execute("for file in `ls -t %s`; do cat $file | grep '<accuracy>' -A 1 | tail -n 1; done" % path))
    time = linesToArray(execute("for file in `ls -t %s`; do cat $file | grep '<wallclock>' -A 1 | tail -n 1; done" % path))
    (time, acc) = zip(*sorted(zip(time, acc), key=lambda x : x[0]))
    return (list(time), list(acc))
```

```python
# cond-ent, ising-disagree, baseline. 
plot(*loadAcc('test/ising/lockdown/default/1/tc99999_c*/policy.xml'))
plot(*loadAcc('test/ising/lockdown/initIID/1/tc99999_c*/policy.xml'))
plot(*loadAcc('test/ising/gibbs/default/0/tc99999_T*/policy.xml'))
legend(['cond-ent + disagree', 'cond-ent only', 'baseline'], loc=4)
```

```python
# cond-ent, ising-disagree, baseline. in wallclock.
plot(*loadAccWallclock('test/ising/lockdown/default/1/tc99999_c*/policy.xml'))
plot(*loadAccWallclock('test/ising/lockdown/initIID/1/tc99999_c*/policy.xml'))
plot(*loadAccWallclock('test/ising/gibbs/default/0/tc99999_T*/policy.xml'))
legend(['cond-ent + disagree', 'cond-ent only', 'baseline'], loc=4)
```
