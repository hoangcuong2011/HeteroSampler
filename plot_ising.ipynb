```python
%pylab inline 
%load_ext autoreload
%autoreload 2
from lxml import etree as Etree
import matplotlib
import matplotlib.pyplot as plt
import sys
import numpy as np
```

```python
# visulize the performance of the training algorithm.
def parseBlock(block):
    block = block.split('\t')
    img = [int(b.split('/')[0]) for b in block if b != ""]
    tag = [int(b.split('/')[1]) for b in block if b != ""]
    return img, tag

def visualizeExample(root, path):
    truth = root.findall(path+'/truth')[0].text
    tag = root.findall(path+'/tag')[0].text
    tag = tag.split('\n')[1]
    headline = truth.split('\n')[1]
    headline = headline.split(' ')
    H = int(headline[0])
    W = int(headline[1])
    (img, tag) = parseBlock(tag)
    (img, gt) = parseBlock(truth.split('\n')[2])
    gt = np.reshape(gt, (H, W), order='F')
    img = np.reshape(img, (H, W), order='F')
    tag = np.reshape(tag, (H, W), order='F')
    print 'image = '
    imshow(img)
    show()
    print 'tag = '
    imshow(tag)
    show()
    print 'truth = '
    imshow(gt)
    show()
name = 'result/ising.xml'
tree = Etree.parse(name)
root = tree.getroot()
pass_s = 'pass_2'
visualizeExample(root, pass_s + '/test/example_28')
score = root.findall(pass_s + '/score')
print score[-1].text
```

```python
# visualize mask of adaptive inference algorithm.
def visualizeMask(mask):
    mask = mask.split('\n')[1]
    mask = mask.split('\t')
    mask = [int(mk) for mk in mask if mk != '']    
    mask = np.reshape(mask, (128, 128), order='F')
    imshow(mask)
    show()

name = 'test/ising/lockdown/default/0/tc99999_c-1.3540563321898453/policy.xml'
tree = Etree.parse(name)
root = tree.getroot()
mask = root.findall('test/example/example_0/mask')[0].text
visualizeMask(mask)
mask = root.findall('test/example/example_1/mask')[0].text
visualizeMask(mask)
mask = root.findall('test/example/example_2/mask')[0].text
visualizeMask(mask)
```
