```python
%pylab inline 
%load_ext autoreload
%autoreload 2
from lxml import etree as Etree
import matplotlib
import matplotlib.pyplot as plt
import sys
import numpy as np
```

```python
def parseBlock(block):
    block = block.split('\t')
    img = [int(b.split('/')[0]) for b in block if b != ""]
    tag = [int(b.split('/')[1]) for b in block if b != ""]
    return img, tag
```

```python
name = 'result/ising.xml'
tree = Etree.parse(name)
root = tree.getroot()
truth = root.findall('pass_0/test/example_0/truth')[0].text
headline = truth.split('\n')[1]
headline = headline.split(' ')
H = int(headline[0])
W = int(headline[1])
(img, gt) = parseBlock(truth.split('\n')[2])
img = np.reshape(img, (H, W), order='F')
gt = np.reshape(gt, (H, W), order='F')
imshow(gt)
show()
```
