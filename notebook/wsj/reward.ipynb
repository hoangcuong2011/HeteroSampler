```python
%pylab inline 
%load_ext autoreload
%autoreload 2
from lxml import etree as Etree
import matplotlib
import matplotlib.pyplot as plt
import sys
import numpy as np
import subprocess

maindir = '../../'
def parse(path):
  tree = Etree.parse(maindir+path)
  root = tree.getroot()
  nodes = root.findall('train/policy_example/*')
  res = list()
  for node in nodes:
    dic = dict()
    for item in node:
      attr = item.attrib
      dic[item.tag + '/' + attr['name']] = float(attr['value'])
    res.append(dic)
  return res

def plotRewardDecay(path):
  files = execute('ls %s' % path).split('\n')[:-1]
  x = []
  y = []
  for f in files:
    tree = Etree.parse(f)
    root = tree.getroot()
    nodes = root.findall('./test/example/*/feat/oracle')
    oracle = [[float(token.split('/')[0]) for token in node.text.replace('\n', '').split('\t') if token != ''] for node in nodes]
    reward = []
    for li in oracle:
      reward += li
    y += reward
  #   x += [float(f.split('tcall_b')[1].split('/')[0])] * len(reward)
    x += [float(root.findall('./test/time')[0].text)] * len(reward)
  plot(x, y, 'r.')

def linesToArray(lines):
    lines = lines.split('\n')
    return [float(line) for line in lines if line != '']
  
def loadAcc(path):
    acc = linesToArray(execute("for file in `ls -t %s`; do cat $file | grep '<accuracy>' -A 1 | tail -n 1; done" % path))
    time = linesToArray(execute("for file in `ls -t %s`; do cat $file | grep '<time>' -A 1 | tail -n 1; done" % path))
    (time, acc) = zip(*sorted(zip(time, acc), key=lambda x : x[0]))
    return (list(time), list(acc))

def loadAccWallclock(path):
    acc = linesToArray(execute("for file in `ls -t %s`; do cat $file | grep '<accuracy>' -A 1 | tail -n 1; done" % path))
    time = linesToArray(execute("for file in `ls -t %s`; do cat $file | grep '<wallclock>' -A 1 | tail -n 1; done" % path))
    (time, acc) = zip(*sorted(zip(time, acc), key=lambda x : x[0]))
    return (list(time), list(acc))

def loadWallclock(path):
    time = linesToArray(execute("for file in `ls -t %s`; do cat $file | grep '<wallclock>' -A 1 | tail -n 1; done" % path))
    sample = linesToArray(execute("for file in `ls -t %s`; do cat $file | grep '<wallclock_sample>' -A 1 | tail -n 1; done" % path))
    policy = linesToArray(execute("for file in `ls -t %s`; do cat $file | grep '<wallclock_policy>' -A 1 | tail -n 1; done" % path))
    return (list(time), list(sample), list(policy))

def plotWallclock(path):
    (time, sample, policy) = loadWallclock(path)
    plot(time, sample)
    plot(time, policy)
    legend(["full-model", "policy"], loc=4)
    title("wallclock")
    xlabel("Average number of transitions per sample.")
    ylabel("wallclock time")
    
def plotSampleDistribution(path):
    tree = Etree.parse(path)
    root = tree.getroot()
    nodes = root.findall('./test/example/*/mask')
    masks = list()
    for node in nodes:
        masks += [int(x) for x in node.text.replace('\n', '').split('\t') if x != '']
    plt.hist(masks, bins=30)
    plt.title('Histogram')
    plt.xlabel('Average #Transition per Sample')
    
plotAcc = lambda name : plot(*loadAcc(name))
extract = lambda dic_l, name : [dic[name] for dic in dic_l]
```

```python
%clear
res = parse("test/wsj/f4/blockpolicy/all/0/tcall_train/policy.xml")
keys = res[0].keys()
if 'item/reward' in keys:
  reward = extract(res, 'item/reward')
if 'item/resp' in keys:
  resp = extract(res, 'item/resp')
if 'feat/oracle-stale' in keys:
  oracle_stale = extract(res, 'feat/oracle-stale')
if 'feat/oracle' in keys:
  oracle = extract(res, 'feat/oracle')
if 'feat/oracle-ent' in keys:
  oracle_ent = extract(res, 'feat/oracle-ent')
if 'feat/cond-lhood' in keys:
  cond_lhood = extract(res, 'feat/cond-lhood')
if 'feat/nb-vary' in keys:
  nb_vary = extract(res, 'feat/nb-vary')
if 'feat/b' in keys:
  bias = extract(res, 'feat/b')
if 'feat/sp' in keys:
  num_sp = extract(res, 'feat/sp')
if 'feat/log-sp' in keys:
  log_num_sp = extract(res, 'feat/log-sp')
if 'feat/cond-ent' in keys:
  cond_ent = extract(res, 'feat/cond-ent')
if 'feat/nb-vary' in keys:
  nb_vary = extract(res, 'feat/nb-vary')
# old_param = zip(extract(res, 'param/b'), extract(res, 'param/cond-ent'), extract(res, 'param/sp'))
```
