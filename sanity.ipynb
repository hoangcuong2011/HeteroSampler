```python
%pylab inline 
%load_ext autoreload
%autoreload 2
from lxml import etree as Etree
import matplotlib
import matplotlib.pyplot as plt
import sys
import numpy as np
import subprocess

# accuracy vs. time.
def execute(cmd):
    ps = subprocess.Popen(cmd, shell=True,stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    output = ps.communicate()[0]
    return output

def linesToArray(lines):
    lines = lines.split('\n')
    return [float(line) for line in lines if line != '']
    
def loadAcc(path):
    acc = linesToArray(execute("for file in `ls -t %s`; do cat $file | grep '<accuracy>' -A 1 | tail -n 1; done" % path))
    time = linesToArray(execute("for file in `ls -t %s`; do cat $file | grep '<time>' -A 1 | tail -n 1; done" % path))
    (time, acc) = zip(*sorted(zip(time, acc), key=lambda x : x[0]))
    return (list(time), list(acc))

def loadAccWallclock(path):
    acc = linesToArray(execute("for file in `ls -t %s`; do cat $file | grep '<accuracy>' -A 1 | tail -n 1; done" % path))
    time = linesToArray(execute("for file in `ls -t %s`; do cat $file | grep '<wallclock>' -A 1 | tail -n 1; done" % path))
    (time, acc) = zip(*sorted(zip(time, acc), key=lambda x : x[0]))
    return (list(time), list(acc))

def loadWallclock(path):
    time = linesToArray(execute("for file in `ls -t %s`; do cat $file | grep '<wallclock>' -A 1 | tail -n 1; done" % path))
    sample = linesToArray(execute("for file in `ls -t %s`; do cat $file | grep '<wallclock_sample>' -A 1 | tail -n 1; done" % path))
    policy = linesToArray(execute("for file in `ls -t %s`; do cat $file | grep '<wallclock_policy>' -A 1 | tail -n 1; done" % path))
    return (list(time), list(sample), list(policy))

def plotWallclock(path):
    (time, sample, policy) = loadWallclock(path)
    plot(time, sample)
    plot(time, policy)
    legend(["full-model", "policy"], loc=4)
    title("wallclock")
    xlabel("Average number of transitions per sample.")
    ylabel("wallclock time")
    
def plotSampleDistribution(path):
    tree = Etree.parse(path)
    root = tree.getroot()
    nodes = root.findall('./test/example/*/mask')
    masks = list()
    for node in nodes:
        masks += [int(x) for x in node.text.replace('\n', '').split('\t') if x != '']
    plt.hist(masks, bins=30)
    plt.title('Histogram')
    plt.xlabel('Average #Transition per Sample')
    
plotAcc = lambda name : plot(*loadAcc(name))
```

```python
# NER-f2
figure(num=None, figsize=(8, 6), dpi=80)
plot(*loadAcc('test/ner/w2/f2/blockpolicy/sanity/0/tcall_b*/policy.xml'))
plot(*loadAcc('test/ner/w2/f2/gibbs/default/0/tcall_T*/policy.xml'))
title('Effect of All Features')
xlabel('Average Transations Per Example')
ylabel('F1')
legend(['all features', 'baseline'], loc=4)
```
