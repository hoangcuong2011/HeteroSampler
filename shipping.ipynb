```python
%pylab inline 
%load_ext autoreload
%autoreload 2
from lxml import etree as Etree
import matplotlib
import matplotlib.pyplot as plt
import sys
import numpy as np
import subprocess

# accuracy vs. time.
def execute(cmd):
    ps = subprocess.Popen(cmd, shell=True,stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    output = ps.communicate()[0]
    return output

def linesToArray(lines):
    lines = lines.split('\n')
    return [float(line) for line in lines if line != '']
    
def loadAcc(path):
    acc = linesToArray(execute("for file in `ls -t %s`; do cat $file | grep '<accuracy>' -A 1 | tail -n 1; done" % path))
    time = linesToArray(execute("for file in `ls -t %s`; do cat $file | grep '<time>' -A 1 | tail -n 1; done" % path))
    (time, acc) = zip(*sorted(zip(time, acc), key=lambda x : x[0]))
    return (list(time), list(acc))

def loadAccWallclock(path):
    acc = linesToArray(execute("for file in `ls -t %s`; do cat $file | grep '<accuracy>' -A 1 | tail -n 1; done" % path))
    time = linesToArray(execute("for file in `ls -t %s`; do cat $file | grep '<wallclock>' -A 1 | tail -n 1; done" % path))
    (time, acc) = zip(*sorted(zip(time, acc), key=lambda x : x[0]))
    return (list(time), list(acc))

def loadWallclock(path):
    time = linesToArray(execute("for file in `ls -t %s`; do cat $file | grep '<wallclock>' -A 1 | tail -n 1; done" % path))
    sample = linesToArray(execute("for file in `ls -t %s`; do cat $file | grep '<wallclock_sample>' -A 1 | tail -n 1; done" % path))
    policy = linesToArray(execute("for file in `ls -t %s`; do cat $file | grep '<wallclock_policy>' -A 1 | tail -n 1; done" % path))
    return (list(time), list(sample), list(policy))

def plotWallclock(path):
    _linewidth = 2.0
    (time, sample, policy) = loadWallclock(path)
    plot(time, sample)
    plot(time, policy)
    legend(["full-model", "policy"], loc=2)
    title("wallclock")
    xlabel("Average number of transitions per sample.")
    ylabel("wallclock time")
    
def plotSampleDistribution(path):
    tree = Etree.parse(path)
    root = tree.getroot()
    nodes = root.findall('./test/example/*/mask')
    masks = list()
    for node in nodes:
        masks += [int(x) for x in node.text.replace('\n', '').split('\t') if x != '']
    plt.hist(masks, bins=30)
    plt.title('Histogram')
    plt.xlabel('Average #Transition per Sample')
    
plotAcc = lambda name : plot(*loadAcc(name))
```

```python
import matplotlib.font_manager as fm
fm.FontProperties(family='font.serif')

def plotPerformanceWithBudgets(paths, _title, _xlabel, _ylabel,
    _linewidth = 3.0,
    _labelsize = 30,
    _legendsize = 25,
    _ticksize = 25,
    _markersize = 0,
    time_scale = 1,
    acc_scale = 1):
    h = figure(num=None, figsize=(10, 7.5), dpi=80)
    (time, acc) = loadAcc(paths[0])
    plot(np.array(time) / time_scale, np.array(acc) / acc_scale, color='r', marker='s', markersize=_markersize, linewidth=_linewidth)
    (time, acc) = loadAcc(paths[1])
    plot(np.array(time) / time_scale, np.array(acc) / acc_scale, color='b', marker='^', markersize=_markersize, linewidth=_linewidth)
    title(_title, fontsize=_labelsize)
    xlabel(_xlabel, fontsize=_labelsize)
    ylabel(_ylabel, fontsize=_labelsize)
    legend(['All features', 'baseline'], loc=4, fontsize=_legendsize)
    tick_params(axis='both', which='major', labelsize=_ticksize)
    pylab.rcParams['xtick.major.pad']='12'
    pylab.rcParams['ytick.major.pad']='12'
    return h
    
def plotPolicyOverhead(path,  _title, _xlabel, _ylabel, 
    _linewidth = 3.0,
    _labelsize = 30,
    _legendsize = 25,
    _ticksize = 25,
    _markersize = 5):
    figure(num=None, figsize=(12, 9), dpi=80)
    (time, sample, policy) = loadWallclock(path)
    plot(time, np.array(policy), color='b', linewidth=_linewidth)
    plot(time, np.array(sample)+np.array(policy), color='r', linewidth=_linewidth, fillstyle='bottom')
    fill_between(time,np.array(sample)+np.array(policy),0,color='#FAA2A2')
    fill_between(time,np.array(policy),0,color='#ACBBFA')
    title(_title, fontsize=_labelsize)
    xlabel(_xlabel, fontsize=_labelsize)
    ylabel(_ylabel, fontsize=_labelsize)
    pylab.rcParams['xtick.major.pad']='12'
    pylab.rcParams['ytick.major.pad']='12'
    legend(['Policy', 'Overall'], loc=2, fontsize=_legendsize)
    tick_params(axis='both', which='major', labelsize=_ticksize)
```

# Ablation Analysis

```python
def plotAblationAnalysis(paths, _title, _legend):
    figure(num=None, figsize=(16, 6), dpi=80)
    _linewidth = 2
    _labelsize = 20
    _ticksize = 15
    _legendsize = 20
    markers = ['+', '*', 'o', 's', '1', '2', '3', '4']
    linestyle = ['-', '-', '-', '--', '--', '--', '-.', '-.']
    for (pi, path) in enumerate(paths):
        plot(*loadAcc(path), linewidth=_linewidth, marker=markers[pi], ls=linestyle[pi])
    title(_title, fontsize=_labelsize)
    xlabel('Average Transations Per Example', fontsize=_labelsize)
    ylabel('F1', fontsize=_labelsize)
    tick_params(axis='both', which='major', labelsize=_ticksize)
    legend(_legend, loc=4, fontsize=_legendsize)

def doAblationAnalysis(paths):
    num = list()
    min_time = 1e20
    min_acc = 1e20
    max_time = 0
    max_acc = -1e20
    for (pi, path) in enumerate(paths):
        num.append(loadAcc(path))
        time = num[-1][0]
        acc = num[-1][1]
        if time[0] < min_time:
            min_time = time[0]
        if time[-1] > max_time:
            max_time = time[-1]
        min_acc = min(min_acc, acc[0])
        max_acc = max(max_acc, acc[-1])
    converge_time = 0
    for n in num:
        time = n[0]
        acc = n[1]
        for i in range(len(time)-1):
            if acc[i+1]-acc[i] < 1e-3 * abs(acc[i]): # converged.
                if time[i] > converge_time:
                    converge_time = time[i]
                break
                
    print 'converge', converge_time
    area = list()
    area_rec = (converge_time - min_time) * (max_acc - min_acc)
    for n in num:
        time = n[0]
        acc = n[1]
        a = 0
        time[0] = min_time
        time[-1] = max_time
        for i in range(1, len(time)):
            if time[i] > converge_time:
                break
            a += .5 * (time[i] - time[i-1]) * (acc[i] + acc[i-1] - 2 * min_acc) # normalize by the first
        area.append(a / area_rec)
    return area

            
        
    
 
```
