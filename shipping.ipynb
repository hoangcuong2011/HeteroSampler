```python
%pylab inline 
%load_ext autoreload
%autoreload 2
from lxml import etree as Etree
import matplotlib
import matplotlib.pyplot as plt
import sys
import numpy as np
import subprocess

# accuracy vs. time.
def execute(cmd):
    ps = subprocess.Popen(cmd, shell=True,stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    output = ps.communicate()[0]
    return output

def linesToArray(lines):
    lines = lines.split('\n')
    return [float(line) for line in lines if line != '']
    
def loadAcc(path):
    acc = linesToArray(execute("for file in `ls -t %s`; do cat $file | grep '<accuracy>' -A 1 | tail -n 1; done" % path))
    time = linesToArray(execute("for file in `ls -t %s`; do cat $file | grep '<time>' -A 1 | tail -n 1; done" % path))
    (time, acc) = zip(*sorted(zip(time, acc), key=lambda x : x[0]))
    return (list(time), list(acc))

def loadAccWallclock(path):
    acc = linesToArray(execute("for file in `ls -t %s`; do cat $file | grep '<accuracy>' -A 1 | tail -n 1; done" % path))
    time = linesToArray(execute("for file in `ls -t %s`; do cat $file | grep '<wallclock>' -A 1 | tail -n 1; done" % path))
    (time, acc) = zip(*sorted(zip(time, acc), key=lambda x : x[0]))
    return (list(time), list(acc))

def loadWallclock(path):
    time = linesToArray(execute("for file in `ls -t %s`; do cat $file | grep '<wallclock>' -A 1 | tail -n 1; done" % path))
    sample = linesToArray(execute("for file in `ls -t %s`; do cat $file | grep '<wallclock_sample>' -A 1 | tail -n 1; done" % path))
    policy = linesToArray(execute("for file in `ls -t %s`; do cat $file | grep '<wallclock_policy>' -A 1 | tail -n 1; done" % path))
    return (list(time), list(sample), list(policy))

def plotWallclock(path):
    (time, sample, policy) = loadWallclock(path)
    plot(time, sample)
    plot(time, policy)
    legend(["full-model", "policy"], loc=2)
    title("wallclock")
    xlabel("Average number of transitions per sample.")
    ylabel("wallclock time")
    
def plotSampleDistribution(path):
    tree = Etree.parse(path)
    root = tree.getroot()
    nodes = root.findall('./test/example/*/mask')
    masks = list()
    for node in nodes:
        masks += [int(x) for x in node.text.replace('\n', '').split('\t') if x != '']
    plt.hist(masks, bins=30)
    plt.title('Histogram')
    plt.xlabel('Average #Transition per Sample')
    
plotAcc = lambda name : plot(*loadAcc(name))
```

# Performance with different budgets

```python
def plotPerformanceWithBudgets(paths, _title, _xlabel, _ylabel,
    _linewidth = 2.0,
    _labelsize = 30,
    _legendsize = 25,
    _ticksize = 25,
    _markersize = 5):
    h = figure(num=None, figsize=(10, 7.5), dpi=80)
    plot(*loadAcc(paths[0]), color='r', marker='s', markersize=_markersize, linewidth=_linewidth)
    # plot(*loadAcc('test/ner/w2/f2/blockpolicy/oracle-nosp/0/tcall_b*/policy.xml'))
    plot(*loadAcc(paths[1]), color='b', marker='^', markersize=_markersize, linewidth=_linewidth)
    title(_title, fontsize=_labelsize)
    xlabel(_xlabel, fontsize=_labelsize)
    ylabel(_ylabel, fontsize=_labelsize)
    legend(['All features', 'baseline'], loc=4, fontsize=_legendsize)
    tick_params(axis='both', which='major', labelsize=_ticksize)
    return h
    
def plotPolicyOverhead(path,  _title, _xlabel, _ylabel, 
    _linewidth = 2.0,
    _labelsize = 30,
    _legendsize = 25,
    _ticksize = 25,
    _markersize = 5):
    figure(num=None, figsize=(12, 9), dpi=80)
    (time, sample, policy) = loadWallclock(path)
    plot(time, policy, color='b', linewidth=_linewidth)
    plot(time, sample, color='r', linewidth=_linewidth)
    title(_title, fontsize=_labelsize)
    xlabel(_xlabel, fontsize=_labelsize)
    ylabel(_ylabel, fontsize=_labelsize)
    legend(['Policy', 'Sampling'], loc=2, fontsize=_legendsize)
    tick_params(axis='both', which='major', labelsize=_ticksize)
```
